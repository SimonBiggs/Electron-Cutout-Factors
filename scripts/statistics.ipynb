```python
def widthRatioMethod(widthTest,ratioTest,widthData,ratioData,factorData):
    
    widthMirrored = append(widthData,widthData)
    ratioLogMirrored = log2(append(ratioData,1/ratioData))
    factorMirrored = append(factorData,factorData)

    widthRatioFit = SmoothBivariateSpline(widthMirrored,ratioLogMirrored,factorMirrored)

    return widthRatioFit.ev(widthTest,log2(ratioTest))
```

## Bound Tests

```python
def univariateGap(xTest,xData):
    xTest = array(xTest)
    return (xTest > min(xData)) & (xTest < max(xData))

def single_univariateGive(xTest,xData,yData):
    initialPredicted = UnivariateSpline(xData,yData)(xTest)

    adjXData = append(xData,xTest)

    negAdjYData = append(yData,initialPredicted-1)
    posAdjYData = append(yData,initialPredicted+1)

    negFitPredicted = UnivariateSpline(adjXData,negAdjYData)(xTest)
    posFitPredicted = UnivariateSpline(adjXData,posAdjYData)(xTest)

    posGive = posFitPredicted - initialPredicted
    negGive = initialPredicted - negFitPredicted

    return mean([posGive,negGive])

def univariateGive(xTest,xData,yData):
    
    dim = shape(xTest)
    
    if size(dim) == 0:
        return single_univariateGive(xTest,xData,yData)
    
    give = zeros(dim)
    
    if size(dim) == 1:
        for i in range(dim[0]):
            give[i] = single_univariateGive(xTest[i],xData,yData)
            
        return give
    
    
    for i in range(dim[0]):
        for j in range (dim[1]):
            give[i,j] = single_univariateGive(xTest[i,j],xData,yData)

    return give
```

## With Bound Tests

```python
widthMethodDisagreement = zeros(size(width))

for i in range(len(width)):
    
    widthTest = width[i]
    factorTest = factor[i]
    
    adjWidthData = delete(width,i)
    adjFactorData = delete(factor,i)
    
    
    if ( (univariateGap(widthTest,adjWidthData)) & 
         (univariateGive(widthTest,adjWidthData,adjFactorData) < parameters['giveThreshold'])):
        
        predictedFactor = widthMethod(widthTest,adjWidthData,adjFactorData)
        widthMethodDisagreement[i] = factorTest - predictedFactor
        
    else:
        widthMethodDisagreement[i] = nan   
    

widthMethodDisagreement = widthMethodDisagreement[~isnan(widthMethodDisagreement)]
std(widthMethodDisagreement)
```

```python
mean(widthMethodDisagreement)
```

```python
t, prob = ttest_1samp(widthMethodDisagreement,0)
prob
```

```python
size(widthMethodDisagreement)
```

```python
widthRatioMethodDisagreement = zeros(size(width))

for i in range(len(width)):
    
    widthTest = width[i]
    ratioTest = ratio[i]
    factorTest = factor[i]
    
    adjWidthData = delete(width,i)
    adjRatioData = delete(ratio,i)
    adjFactorData = delete(factor,i)
    
    
    adjWidthMirrored = append(adjWidthData,adjWidthData)
    adjRatioLogMirrored = log2(append(adjRatioData,1/adjRatioData))
    adjFactorMirrored = append(adjFactorData,adjFactorData)
    
    gap = angle_gap(widthTest,log2(ratioTest),adjWidthMirrored,adjRatioLogMirrored,1,2)
    give = fit_give(widthTest,log2(ratioTest),adjWidthMirrored,adjRatioLogMirrored,adjFactorMirrored)
    
    if (gap < parameters['gapThreshold']) & (give < parameters['giveThreshold']):
        predictedFactor = widthRatioMethod(widthTest,ratioTest,adjWidthData,adjRatioData,adjFactorData)
        widthRatioMethodDisagreement[i] = factorTest - predictedFactor
        
    else:
        widthRatioMethodDisagreement[i] = nan

widthRatioMethodDisagreement = widthRatioMethodDisagreement[~isnan(widthRatioMethodDisagreement)]
std(widthRatioMethodDisagreement)
```

```python
mean(widthRatioMethodDisagreement)
```

```python
t, prob = ttest_1samp(widthRatioMethodDisagreement,0)
prob
```
